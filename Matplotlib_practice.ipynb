```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize = (5,3), dpi = 100)

x2 = np.arange(0, 4.5, 0.5)
x2
plt.plot(x2[:6], x2[:6]**2)
plt.plot(x2[5:], x2[5:]**2)


plt.plot([2,4,6,7], [3,8,7,10], label = 'Regress', color = 'y', linestyle = '--', linewidth = 4, marker= '*', markersize = 10, markeredgecolor  = 'r')
plt.title("First Graph", fontdict = {'fontname': 'Comic Sans MS', 'fontsize' : 10})
plt.xlabel("X-Axis", fontdict= {'fontname': 'Comic Sans MS', 'fontsize' : 10})
plt.ylabel("Y-Axis", fontdict = {'fontname': 'Comic Sans MS', 'fontsize' : 10})
plt.xticks([2, 4, 6, 8])
plt.yticks([2, 4, 6, 8, 10])
plt.legend()
plt.savefig("First_graph.png")
plt.show()
```

```python
labels = ['a', 'b', 'c']
values = [4, 8, 16]
bars = plt.bar(labels, values, color = '#ababab')
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.show()
```

```python
plt.plot([2,4,5,6], [3,6,8,10],'ro')
plt.axis([0,8,0,20]) # axi== we give xmin xmax and ymin and ymax
plt.show()
```

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)
plt.scatter(t, t**2)
plt.plot(t, t, 'ro', t, t**2, 'bs', t, t**3, 'g^')
plt.grid(True)
plt.show()
```
