```python
missing_values = data.isnull().sum()
```

```python
missing_values[:10]
```

```python
total_cells = np.product(data.shape)
total_missing_values = missing_values.sum()
total_percentage_of_missing_values = (total_missing_values/total_cells)*100
print("Percentage of missing values in the data = ",total_percentage_of_missing_values )
```

```python
data_set = data.loc[:, 'EPA':'Season'].head()
data_set
```

```python
data_set.fillna(0)
```

```python
data_set.fillna(method = 'bfill', axis = 0).fillna(0)
```

```python
data_set.fillna(method = 'ffill', axis = 0).fillna(0)
```

```python
#new_data = data.copy()
from sklearn.impute import SimpleImputer
myimpute = SimpleImputer()
data_impute = myimpute.fit_transform(data_set)
new = pd.DataFrame(data_impute)
new.head()
```

```python
data_set.head()
```

```python
new_data = data_set.copy()

# make new columns indicating what will be imputed
cols_with_missing = (col for col in new_data.columns 
                                 if new_data[col].isnull().any())
for col in cols_with_missing:
    new_data[col + '_was_missing'] = new_data[col].isnull()
```
